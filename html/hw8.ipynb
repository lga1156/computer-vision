```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import urllib.request
import os
```

## Задание 1

Выбрать тестовое изображение в оттенках серого и преобразовать его в бинарное с помощью пороговой сегментации.

```python
img_path = 'test_image.png'
if not os.path.exists(img_path):
    url = 'https://raw.githubusercontent.com/opencv/opencv/master/samples/data/sudoku.png'
    urllib.request.urlretrieve(url, img_path)

img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

ret, binary_img = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Исходное изображение (оттенки серого)')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(binary_img, cmap='gray')
plt.title(f'Бинарное изображение (Порог Оцу: {ret})')
plt.axis('off')

plt.tight_layout()
plt.show()
```

## Задание 2

Выполнить эрозию полученного изображения с помощью круглого примитива размера 5х5 (примитив можно получить функцией cv.getStructuringElement).

```python
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
eroded_img = cv2.erode(binary_img, kernel, iterations=1)
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(binary_img, cmap='gray')
plt.title('Исходное бинарное изображение')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(eroded_img, cmap='gray')
plt.title('После эрозии')
plt.axis('off')

plt.tight_layout()
plt.show()
```

## Задание 3

Выполнить дилатацию изображения с таким же примитивом.

```python
dilated_img = cv2.dilate(binary_img, kernel, iterations=1)
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(binary_img, cmap='gray')
plt.title('Исходное бинарное изображение')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(dilated_img, cmap='gray')
plt.title('После дилатации')
plt.axis('off')

plt.tight_layout()
plt.show()
```

## Задание 4

Выполнить операции размыкания исходного бинарного изображения.

```python
opened_img = cv2.morphologyEx(binary_img, cv2.MORPH_OPEN, kernel)
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(binary_img, cmap='gray')
plt.title('Исходное бинарное изображение')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(opened_img, cmap='gray')
plt.title('После размыкания (Opening)')
plt.axis('off')

plt.tight_layout()
plt.show()
```

## Задание 5

Выполнить операцию замыкания исходного бинарного изображения.

```python
closed_img = cv2.morphologyEx(binary_img, cv2.MORPH_CLOSE, kernel)
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(binary_img, cmap='gray')
plt.title('Исходное бинарное изображение')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(closed_img, cmap='gray')
plt.title('После замыкания (Closing)')
plt.axis('off')

plt.tight_layout()
plt.show()
```

## Задание 6

Получить изображение, на котором выделены все границы исходного бинарного изображения.

```python
gradient_img = cv2.morphologyEx(binary_img, cv2.MORPH_GRADIENT, kernel)
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(binary_img, cmap='gray')
plt.title('Исходное бинарное изображение')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(gradient_img, cmap='gray')
plt.title('Границы (Морфологический градиент)')
plt.axis('off')

plt.tight_layout()
plt.show()
```
