```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Задание 1

Подготовить тестовое изображение в градациях серого.

```python
img = np.zeros((256, 256), dtype=np.uint8)

for i in range(256):
    img[:, i] = i

cv2.circle(img, (128, 128), 50, 255, -1) # Белый круг
cv2.rectangle(img, (30, 30), (80, 80), 0, -1) # Черный квадрат
cv2.rectangle(img, (170, 170), (220, 220), 128, -1) # Серый квадрат

plt.figure(figsize=(6, 6))
plt.imshow(img, cmap='gray')
plt.title('Тестовое изображение в градациях серого')
plt.axis('off')
plt.show()
```

## Задание 2

Выполнить пороговую сегментацию изображения.

```python
# Задаем порог, равный 127
threshold_value = 127
ret, thresh_img = cv2.threshold(img, threshold_value, 255, cv2.THRESH_BINARY)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Исходное изображение')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(thresh_img, cmap='gray')
plt.title(f'Пороговая сегментация (порог = {threshold_value})')
plt.axis('off')

plt.tight_layout()
plt.show()
```

## Задание 3

Выполнить двухпороговую сегментацию изображения.

```python
lower_threshold = 100
upper_threshold = 200

double_thresh_img = cv2.inRange(img, lower_threshold, upper_threshold)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Исходное изображение')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(double_thresh_img, cmap='gray')
plt.title(f'Двухпороговая сегментация\n({lower_threshold} <= Интенсивность <= {upper_threshold})')
plt.axis('off')

plt.tight_layout()
plt.show()
```

## Задание 4

Выполнить сегментацию изображения с глобальным порогом.

```python
global_threshold = int(np.mean(img))

ret, global_thresh_img = cv2.threshold(img, global_threshold, 255, cv2.THRESH_BINARY)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Исходное изображение')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(global_thresh_img, cmap='gray')
plt.title(f'Глобальный порог (T = {global_threshold})')
plt.axis('off')

plt.tight_layout()
plt.show()
```

## Задание 5

Выполнить сегментацию изображения методом Оцу.

```python
otsu_threshold, otsu_thresh_img = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Исходное изображение')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(otsu_thresh_img, cmap='gray')
plt.title(f'Сегментация методом Оцу\n(Подобранный порог: {otsu_threshold})')
plt.axis('off')

plt.tight_layout()
plt.show()
```
