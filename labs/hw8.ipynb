```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import urllib.request
import os
```

## Задание 1

Выбрать тестовое изображение в оттенках серого и преобразовать его в бинарное с помощью пороговой сегментации.

```python
# Скачиваем тестовое изображение, если его еще нет
img_path = 'test_image.png'
if not os.path.exists(img_path):
    url = 'https://raw.githubusercontent.com/opencv/opencv/master/samples/data/sudoku.png'
    urllib.request.urlretrieve(url, img_path)

# 1. Загружаем изображение в оттенках серого
img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

# 2. Преобразуем в бинарное с помощью пороговой сегментации (глобальный порог + метод Оцу для автоматического выбора порога)
ret, binary_img = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# 3. Отображаем результат
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Исходное изображение (оттенки серого)')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(binary_img, cmap='gray')
plt.title(f'Бинарное изображение (Порог Оцу: {ret})')
plt.axis('off')

plt.tight_layout()
plt.show()
```

## Задание 2

Выполнить эрозию полученного изображения с помощью круглого примитива размера 5х5 (примитив можно получить функцией cv.getStructuringElement).

```python
# Создаем круглый примитив (структурирующий элемент) размера 5x5
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))

# Выполняем эрозию
eroded_img = cv2.erode(binary_img, kernel, iterations=1)

# Отображаем результат
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(binary_img, cmap='gray')
plt.title('Исходное бинарное изображение')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(eroded_img, cmap='gray')
plt.title('После эрозии')
plt.axis('off')

plt.tight_layout()
plt.show()
```

## Задание 3

Выполнить дилатацию изображения с таким же примитивом.

```python
# Выполняем дилатацию с тем же примитивом (kernel)
dilated_img = cv2.dilate(binary_img, kernel, iterations=1)

# Отображаем результат
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(binary_img, cmap='gray')
plt.title('Исходное бинарное изображение')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(dilated_img, cmap='gray')
plt.title('После дилатации')
plt.axis('off')

plt.tight_layout()
plt.show()
```

## Задание 4

Выполнить операции размыкания исходного бинарного изображения.

```python
# Выполняем операцию размыкания (Opening)
# Размыкание - это эрозия, за которой следует дилатация. Используем тот же примитив.
opened_img = cv2.morphologyEx(binary_img, cv2.MORPH_OPEN, kernel)

# Отображаем результат
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(binary_img, cmap='gray')
plt.title('Исходное бинарное изображение')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(opened_img, cmap='gray')
plt.title('После размыкания (Opening)')
plt.axis('off')

plt.tight_layout()
plt.show()
```

## Задание 5

Выполнить операцию замыкания исходного бинарного изображения.

```python
# Выполняем операцию замыкания (Closing)
# Замыкание - это дилатация, за которой следует эрозия. Используем тот же примитив.
closed_img = cv2.morphologyEx(binary_img, cv2.MORPH_CLOSE, kernel)

# Отображаем результат
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(binary_img, cmap='gray')
plt.title('Исходное бинарное изображение')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(closed_img, cmap='gray')
plt.title('После замыкания (Closing)')
plt.axis('off')

plt.tight_layout()
plt.show()
```

## Задание 6

Получить изображение, на котором выделены все границы исходного бинарного изображения.

```python
# Получаем границы изображения с помощью морфологического градиента
# Морфологический градиент - это разность между дилатацией и эрозией изображения.
gradient_img = cv2.morphologyEx(binary_img, cv2.MORPH_GRADIENT, kernel)

# Альтернативный способ: вычитание эрозии из исходного изображения (внутренние границы)
# boundary_img = cv2.subtract(binary_img, eroded_img)

# Отображаем результат
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(binary_img, cmap='gray')
plt.title('Исходное бинарное изображение')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(gradient_img, cmap='gray')
plt.title('Границы (Морфологический градиент)')
plt.axis('off')

plt.tight_layout()
plt.show()
```
